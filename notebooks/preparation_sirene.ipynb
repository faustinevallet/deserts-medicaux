```python
df = etab_df.merge(entities_df, on='siren', how='outer', indicator=True)
```

```python
df = df[df.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE) | df.activiteprincipaleunitelegale.str.startswith(ACTIVITY_CODE)] # 86.
```

```python
df.shape
```

```python
ps_df = (df.loc[df.activiteprincipaleetablissement.str.startswith(ACTIVITY_CODE) | df.activiteprincipaleunitelegale.str.startswith(ACTIVITY_CODE)]
         .assign(etab_la = lambda df: df.activiteprincipaleetablissement.str.slice(0,2))
         .assign(ent_la = lambda df: df.activiteprincipaleunitelegale.str.slice(0,2))
         .assign(cross_la = lambda df: df.etab_la + '_' + df.ent_la)
        )
```
